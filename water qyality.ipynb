```python
# visulaization dataset also checking for outliers
fig, ax = plt.subplots(ncols=5 ,nrows=2, figsize = (20,10))
ax = ax.flatten()
index = 0
for col,values in df.items():
    sns.boxplot(y=col,data=df,ax=ax[index])
    index += 1
```
